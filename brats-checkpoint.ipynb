```python
data_root = '/kaggle/input/bratstask-1/BraTS2025-GLI-PRE-Challenge-TrainingData/'
target_shape = (48, 48, 48)

def preprocess_and_save_case(case_dir, target_shape, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    modalities = ['t1n', 't1c', 't2w', 't2f']
    images = []
    files = os.listdir(case_dir)
    for mod in modalities:
        file_name = [f for f in files if f.endswith(f'-{mod}.nii')][0]
        img = nib.load(os.path.join(case_dir, file_name)).get_fdata()
        images.append(img)
    image = np.stack(images, axis=-1)
    mask_file = [f for f in files if f.endswith('-seg.nii')][0]
    mask = nib.load(os.path.join(case_dir, mask_file)).get_fdata()
    mask_remapped = mask.copy()
    mask_remapped[mask == 2] = 1
    mask_remapped[mask == 3] = 2

    # Z-score normalization (per modality, only brain)
    image_norm = image.copy()
    for i in range(image.shape[-1]):
        modality = image[..., i]
        brain_mask = modality > 0
        mean = modality[brain_mask].mean() if np.any(brain_mask) else 0
        std = modality[brain_mask].std() if np.any(brain_mask) else 1
        image_norm[..., i][brain_mask] = (modality[brain_mask] - mean) / (std + 1e-8)

    # Resize to target shape
    factors = [t / s for t, s in zip(target_shape, image_norm.shape[:3])]
    resized_image = np.zeros(target_shape + (image_norm.shape[3],), dtype=np.float32)
    for i in range(image_norm.shape[3]):
        resized_image[..., i] = zoom(image_norm[..., i], factors, order=1)
    resized_mask = zoom(mask_remapped, factors, order=0).astype(np.uint8)

    np.save(os.path.join(save_dir, 'input.npy'), resized_image)
    np.save(os.path.join(save_dir, 'label.npy'), resized_mask)
    return os.path.join(save_dir, 'input.npy'), os.path.join(save_dir, 'label.npy')

# List directories and split
case_dirs = sorted([os.path.join(data_root, d) for d in os.listdir(data_root) if os.path.isdir(os.path.join(data_root, d))])
train_dirs, val_dirs = train_test_split(case_dirs, test_size=0.1, random_state=42)

train_save_root = '/kaggle/working/train'
val_save_root = '/kaggle/working/val'
train_inputs, train_labels, val_inputs, val_labels = [], [], [], []

for case_dir in train_dirs:
    case_name = os.path.basename(case_dir)
    save_dir = os.path.join(train_save_root, case_name)
    input_path, label_path = preprocess_and_save_case(case_dir, target_shape, save_dir)
    train_inputs.append(input_path)
    train_labels.append(label_path)

for case_dir in val_dirs:
    case_name = os.path.basename(case_dir)
    save_dir = os.path.join(val_save_root, case_name)
    input_path, label_path = preprocess_and_save_case(case_dir, target_shape, save_dir)
    val_inputs.append(input_path)
    val_labels.append(label_path)

print("Done preprocessing and saving .npy files.")
```

```python
def conv3d_block(input_tensor, n_filters, kernel_size=3, batchnorm=True):
    x = layers.Conv3D(filters=n_filters, kernel_size=kernel_size,
                      kernel_initializer='he_normal', padding='same')(input_tensor)
    if batchnorm:
        x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv3D(filters=n_filters, kernel_size=kernel_size,
                      kernel_initializer='he_normal', padding='same')(x)
    if batchnorm:
        x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    return x

def build_3d_unet(input_shape=(48,48,48,4), n_filters=32, n_classes=3, dropout=0.3):
    inputs = layers.Input(input_shape)
    c1 = conv3d_block(inputs, n_filters)
    p1 = layers.MaxPooling3D((2, 2, 2))(c1)
    p1 = layers.Dropout(dropout)(p1)
    c2 = conv3d_block(p1, n_filters*2)
    p2 = layers.MaxPooling3D((2, 2, 2))(c2)
    p2 = layers.Dropout(dropout)(p2)
    c3 = conv3d_block(p2, n_filters*4)
    p3 = layers.MaxPooling3D((2, 2, 2))(c3)
    p3 = layers.Dropout(dropout)(p3)
    c4 = conv3d_block(p3, n_filters*8)
    u5 = layers.Conv3DTranspose(n_filters*4, (2,2,2), strides=(2,2,2), padding='same')(c4)
    u5 = layers.concatenate([u5, c3])
    u5 = layers.Dropout(dropout)(u5)
    c5 = conv3d_block(u5, n_filters*4)
    u6 = layers.Conv3DTranspose(n_filters*2, (2,2,2), strides=(2,2,2), padding='same')(c5)
    u6 = layers.concatenate([u6, c2])
    u6 = layers.Dropout(dropout)(u6)
    c6 = conv3d_block(u6, n_filters*2)
    u7 = layers.Conv3DTranspose(n_filters, (2,2,2), strides=(2,2,2), padding='same')(c6)
    u7 = layers.concatenate([u7, c1])
    u7 = layers.Dropout(dropout)(u7)
    c7 = conv3d_block(u7, n_filters)
    outputs = layers.Conv3D(n_classes, (1,1,1), activation='softmax')(c7)
    model = Model(inputs=[inputs], outputs=[outputs])
    return model

def dice_loss(y_true, y_pred, smooth=1e-6):
    axes = tuple(range(1, len(y_true.shape)-1))
    intersection = tf.reduce_sum(y_true * y_pred, axis=axes)
    union = tf.reduce_sum(y_true + y_pred, axis=axes)
    dice = (2. * intersection + smooth) / (union + smooth)
    return 1 - tf.reduce_mean(dice)
```

```python
import numpy as np

def dice_score(y_true, y_pred, smooth=1e-6):
    y_true_f = y_true.flatten()
    y_pred_f = y_pred.flatten()
    intersection = np.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (np.sum(y_true_f) + np.sum(y_pred_f) + smooth)

def iou_score(y_true, y_pred, smooth=1e-6):
    y_true_f = y_true.flatten()
    y_pred_f = y_pred.flatten()
    intersection = np.sum(y_true_f * y_pred_f)
    union = np.sum(y_true_f) + np.sum(y_pred_f) - intersection
    return (intersection + smooth) / (union + smooth)

# To calculate for ALL tumor regions combined (rather than individual classes)
def binary_mask(mask):
    return (mask > 0).astype(np.uint8)

slice_indices = [12, 24, 36]  # Example: same slices you visualized
dice_scores = []
iou_scores = []

for slc in slice_indices:
    true_mask_slice = binary_mask(y_true[:, :, slc])   # Use your arrays here!
    pred_mask_slice = binary_mask(y_pred[:, :, slc])
    dice_scores.append(dice_score(true_mask_slice, pred_mask_slice))
    iou_scores.append(iou_score(true_mask_slice, pred_mask_slice))

print("Dice scores:", dice_scores)
print("IoU scores:", iou_scores)
```

```python
import numpy as np

def dice_score(y_true, y_pred, smooth=1e-6):
    y_true_f = y_true.flatten()
    y_pred_f = y_pred.flatten()
    intersection = np.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (np.sum(y_true_f) + np.sum(y_pred_f) + smooth)

def iou_score(y_true, y_pred, smooth=1e-6):
    y_true_f = y_true.flatten()
    y_pred_f = y_pred.flatten()
    intersection = np.sum(y_true_f * y_pred_f)
    union = np.sum(y_true_f) + np.sum(y_pred_f) - intersection
    return (intersection + smooth) / (union + smooth)

def binary_mask(mask):
    # Combine all tumor regions into one binary mask for global Dice/IoU.
    return (mask > 0).astype(np.uint8)

all_dice_scores = []
all_iou_scores = []

# Loop through the entire validation set
for X_batch, y_batch_true in val_dataset:
    y_batch_pred = model.predict(X_batch)  # Your trained model's predictions
    batch_size = X_batch.shape[0]
    for i in range(batch_size):
        y_true_labels = np.argmax(y_batch_true.numpy()[i], axis=-1)     # shape: (48, 48, 48)
        y_pred_labels = np.argmax(y_batch_pred[i], axis=-1)
        for slc in range(y_true_labels.shape[2]):
            y_true_slice = binary_mask(y_true_labels[:, :, slc])
            y_pred_slice = binary_mask(y_pred_labels[:, :, slc])
            dice = dice_score(y_true_slice, y_pred_slice)
            iou = iou_score(y_true_slice, y_pred_slice)
            all_dice_scores.append(dice)
            all_iou_scores.append(iou)

mean_dice = np.mean(all_dice_scores)
mean_iou = np.mean(all_iou_scores)

print(f"Mean Dice: {mean_dice}")
print(f"Mean IoU: {mean_iou}")
```
